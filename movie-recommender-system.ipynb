```python
# Instead of selecting only a subset of columns, keep all original columns.
# movies = movies[['show_id','title','description','listed_in']] # Comment out this line.

# Alternatively, if you need to drop columns,
# do it after creating the 'tags' column:
movies['tags'] = movies['description'] + movies['listed_in'].str.split(',').apply(lambda x: [i.strip() for i in x])
# Now, drop unnecessary columns if required:
movies = movies[['show_id', 'title', 'description', 'listed_in', 'tags']]
```

```python
movies['tags'] = movies['description'] + movies['listed_in'].str.split(',').apply(lambda x: [i.strip() for i in x]) + movies['cast'].str.split(',').apply(lambda x: [i.strip() for i in x]) + movies['director'].str.split(',').apply(lambda x: [i.strip() for i in x])
```

```python
movies.head()
```

```python
new_df = movies[['show_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][2]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
      y.append(ps.stem(i))
  return " ".join(y)

```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
      print(new_df.iloc[i[0]].title)

```

```python
recommend('Ganglands')
```
